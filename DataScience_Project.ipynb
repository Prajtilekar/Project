# **Exploratory Analysis and Visualization**



```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Calucating MEAN of release_year
netflix_df["release_year"].value_counts().mean()
```

The Netflix Release years, visualized via **HORIZONTAL BAR GRAPH**.

```python
plt.title('Netflix Release Years', fontsize=25)
netflix_df["release_year"].value_counts().plot.barh(figsize=(30,20))
plt.show()
```

The year 2018 is the highest of the Release the Movies and Tv Shows followed by 2017 is 2nd highest.

Visualize the amount of **TV SHOWS VS MOVIES** on Netflix!

```python
# TV SHOW VS MOVIES On NETFLIX - visualised by a BAR CHART
plt.figure(figsize=(10,5))
plt.title('Type of shows on Netflix')
sns.barplot(x=netflix_df['type'], y=netflix_df.index)
plt.show()
```

```python
# The TV Show VS Movies On NETFLIX - visualised by a PIE CHART.
plt.title('Type of shows on Netflix')
netflix_df['type'].value_counts().plot.pie(autopct='%1.1f%%',figsize=(12,8))
plt.show()
```

**Visualize movies and tv shows based on released years**

```python
plt.figure(figsize = (20,10))  #size of the chart

sns.countplot(x="release_year", data=netflix_df, order = netflix_df['release_year'].value_counts(ascending=True)[-20:].index,hue='type')   
plt.title('MOVIES AND TV SHOWS RELEASED YEARS', fontsize=25)     # set title of this chart
plt.ylabel('Top', fontsize=25)                 # set y axis column name
plt.xlabel('release_year',fontsize=25)         # set x axis column name              
plt.show()
```

As per the above plots, the content of movie is greater than TV shows.

**The Netflix Rating Distribution, visualised by a COUNT PLOT.**

Ratings Guide:

TV-MA : Unsuitable for children under 17 (Mature Audience Only).

TV-14 : Unsuitable for children under 14.

TV-PG : Parents or guardians may find inappropriate for younger children.

R : Under 17 requires accompanying parent or adult guardian (Restricted).

PG-13 : Parents strongly cautioned, some material may be inappropriate for children under 13.

TV-Y : Programs aimed at a very young audience, including children from ages 2-6.

TV-Y7 : Programs most appropriate for children age 7 and up.

PG : Some material may not be suitable for children (Parental Guidance suggested).

TV-G : Programs suitable for all ages; these are not necessarily children's shows.

NR : (Not Rated)

G : (General Audiences)

TV-Y7-FV : Programming with fantasy violence that may be more intense or more combative than other programming in the TV-Y7 category.

UR : (Un-rated) (Same as NR)

NC-17 : No children under 17

```python
plt.figure(figsize = (15,8))
plt.title('Netflix Ratings Distribution',fontsize=25)
sns.countplot(x='rating', data = netflix_df)
plt.show()
```

The largest count of Netflix content is made with a TV-MA i.e TV-MA : Unsuitable for children under 17 (Mature Audience Only).

â€œTV-MAâ€ is a rating assigned by the TV Parental Guidelines to a television program designed for mature audiences only.

The **Top 5 countries with Highest Movies/TV Shows**, visualised by a PIE CHART.

```python
plt.figure(figsize=(15,10))
plt.title('Top 5 Netflix Countries',fontsize=20)
colors = sns.color_palette("husl", 9)
netflix_df["country"].value_counts().nlargest(n=5).plot.pie(autopct='%1.1f%%',figsize=(15,10), colors=colors)
plt.show()
```

**Top 5 Netflix country distribution seperated by type of release** (i.e Movies, TV shows), visualizing via Bar Graph.

```python
plt.figure(figsize = (15,8))
plt.title('Top 5 Netflix country distribution seperated by type of release',fontsize=20)
sns.countplot(x='country', data=netflix_df, hue='type', order=netflix_df.country.value_counts().iloc[:5].index)
plt.show()
```

From the above graphs we conclude that US,India,UK,Japan and South Korea are leading countries in Content Creation on Netflix.

**Top 10 Geners being most watched on Netflix:**

```python
plt.figure(figsize = (12,8))
sns.countplot(y='genre',order = netflix_df['genre'].value_counts().index[:10],data = netflix_df)
plt.title('Top 20 Genres on Netflix');
```

The Mostly liked Geners by the Public is Dramas,Comedy and Documentation.
